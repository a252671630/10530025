```python
import numpy as np
import matplotlib.pyplot as plt

class Man:
    def __init__(self,x):
        self.x = np.arange(0,6,0.1)
    def sin(self):
        y1 = np.sin(self.x)
        plt.plot(self.x,y1, label="sin")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.title('sin & cos')
        plt.show()
    def cos(self):
        y2 = np.cos(self.x)
        plt.plot(self.x,y2, label="cos")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.title('sin & cos')
        plt.show()
```

```python
m = Man("")
m.sin()
m.cos()
```
